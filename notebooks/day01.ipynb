```python
import re

def clean_string(s: str) -> str:
    s = s.strip().lower()
    return re.sub(r'\s+', ' ', s)

print(clean_string("   Hoc     py that     vui"))
```

```python
def mean (lst: list[float]) -> float:
    return sum(lst) / len(lst)

print(mean([1,2,3]))
```
